Когда можно использовать urllib.request:
Если тебе нужно использовать исключительно стандартную библиотеку Python и нет возможности устанавливать дополнительные пакеты (например, в ограниченных окружениях).

Для очень простых задач, где дополнительные возможности requests не нужны.

В большинстве случаев, особенно для более сложных задач, requests значительно удобнее и эффективнее. Если нет особых ограничений, рекомендую использовать requests.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

soup = BeautifulSoup(html, 'html.parser'): Создает объект BeautifulSoup из HTML строки html. Второй аргумент 'html.parser' указывает, какой парсер использовать для разбора HTML. html.parser — это встроенный в Python парсер, он достаточно быстрый и обычно подходит для большинства задач. Beautiful Soup поддерживает и другие парсеры (например, lxml, html5lib), которые могут быть более точными или быстрыми в определенных ситуациях.

table = soup.find('table', attrs = {'class' : 'wikitable sortable'}): Ищет в HTML документе (в объекте soup) первый элемент < table> с атрибутом class="wikitable sortable" и сохраняет его в переменную table (в данном коде table не используется).

cnt = 0: Инициализирует счетчик cnt нулем.

for tr in soup.find_all('tr'):: Цикл по всем элементам < tr > (строкам таблицы) в HTML документе. soup.find_all('tr') возвращает список всех найденных элементов < tr >.

cnt += 1: Увеличивает счетчик cnt на 1 для каждой найденной строки таблицы.

for td in tr.find_all(['td', 'th']):: Вложенный цикл по всем элементам < td > (ячейкам таблицы) и < th > (заголовкам таблицы) внутри текущей строки tr. tr.find_all(['td', 'th']) возвращает список всех найденных ячеек и заголовков в текущей строке.

cnt *= 2: Умножает счетчик cnt на 2 для каждой найденной ячейки или заголовка.

print(cnt): Выводит итоговое значение счетчика cnt.

Суть кода:
Код подсчитывает значение cnt по следующему правилу: для каждой строки таблицы cnt увеличивается на 1, а затем для каждой ячейки или заголовка в этой строке cnt умножается на 2. Таким образом, cnt зависит от количества строк и ячеек/заголовков в таблице на указанной странице.

```python
# page = urlopen('https://stepik.org/media/attachments/lesson/245130/6.html') 
page = requests.get('https://stepik.org/media/attachments/lesson/245130/6.html')# скачиваем файл

# Мы можем распечатать HTML-содержимое страницы, используя свойство content:
page.content

# html = page.read().decode('utf8') # считываем содержимое

# soup = BeautifulSoup(html, 'html.parser') 
soup = BeautifulSoup(page.content, 'html.parser') # делаем суп

table = soup.find('table', attrs = {'class' : 'wikitable sortable'})
cnt = 0
for tr in soup.find_all('tr'):
    cnt += 1
    for td in tr.find_all(['td', 'th']):
        cnt *= 2
print(cnt)
```

Парсинг сайта для прогноза погоды

```python
page = requests.get('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168')
soup = BeautifulSoup(page.content, 'html.parser')
seven_day = soup.find(id="seven-day-forecast-container")
forecast_items = seven_day.find_all(class_="tombstone-container")
tonight = forecast_items[0]
print(tonight.prettify())
```

```python
period = tonight.find(class_="period-name").get_text()
period
```

```python
short_desc = tonight.find(class_="short-desc").get_text()
short_desc
```

```python
temp = tonight.find(class_="temp temp-high").get_text()
temp
```

```python
img = tonight.find('img')
desc = img['title']
desc
```

```python
period_tags = seven_day.select('.tombstone-container .period-name')
periods = [pt.get_text() for pt in period_tags]
periods
```

```python
short_descs = [sd.get_text() for sd in seven_day.select(".tombstone-container .short-desc")]
short_descs
```

```python
temps = [t.get_text() for t in seven_day.select(".tombstone-container .temp")]
temps
```

```python
descs = [d['title'] for d in seven_day.select(".tombstone-container img")]
descs
```

```python
weather = pd.DataFrame({
    'period': periods,
    'short_descs': short_descs,
    'temp': temps,
    'desc': descs
})

weather
```

```python
temp_nums = weather['temp'].str.extract(r"(\d+)", expand=False) 
weather["temp_num"] = temp_nums.astype('Int64', errors='ignore')
weather
```

```python
weather["temp_num"].mean()
```

```python
is_night = weather["temp"].str.contains("Low")
weather["is_night"] = is_night
weather
```

```python
weather[is_night]
```
